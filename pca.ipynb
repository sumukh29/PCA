```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/sumukh/Downloads/sample.csv')
data = data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6'], axis=1)
print(data.head())
```

```python
mean = np.mean(data, axis=0)
print("Mean\n")
print(mean)
```

```python
cov_matrix = np.cov(data.T)
print("Covariance Matrix\n")
print(cov_matrix)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)
print("Eigen Values")
print(eigen_values)
print("for lamda 1 eigen vector v1")
print(eigen_vectors[0])
print("for Lamda 2 eigen vector v2")
print(eigen_vectors[1])
print("Normalized eigen vector")
eigen_vectors = eigen_vectors.T
print(eigen_vectors[0])
print(eigen_vectors[1])
```

```python
new_data = np.dot(data, eigen_vectors.T[0])
print("New data set for principal component 1")
print(new_data)
```

```python
theta1 = np.arctan(eigen_vectors[0][1]/eigen_vectors[0][0])
theta2 = np.arctan(eigen_vectors[1][1]/eigen_vectors[1][0])
theta1 = np.rad2deg(theta1)
theta2 = np.rad2deg(theta2)
print("Theta 1 in degree")
print(theta1)
print("Theta 2 in degree")
print(theta2)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(data['A'], data['B'])
plt.plot([0, eigen_vectors[0, 0]], [0, eigen_vectors[1, 0]], color='r')
plt.plot([0, eigen_vectors[0, 1]], [0, eigen_vectors[1, 1]], color='b')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Data with Eigenvectors')
plt.grid()
plt.show()
```
